Data From: https://data.imls.gov/Public-Libraries-Survey/Library-Systems-FY-2015-Public-Libraries-Survey-Ad/ve4v-rafu/data

```python
imls = pd.read_csv('/Users/Peter.Organisciak/Downloads/Library_Systems__FY_2015_Public_Libraries_Survey__Administrative_Entities_Data_.csv')

imls.columns = imls.columns.str.title()

for col in imls.columns:
    if ('Revenue' in col) and (imls[col].dtype != int):
        imls[col] = imls[col].str.replace('$', '').astype('float').astype('int')

imls = imls.rename(columns={'Unduplicated Population': 'Population', 'Total Revenue': 'Revenue'})
imls_subset = imls[['State', 'County', 'Library Name', 'Population', 
                    'Revenue', 'Salaries', 'Mls Librarian Staff']]
imls_by_state = imls_subset.groupby('State').sum()
imls_by_state.to_csv('../data/imls_mini_by_state.csv')
```
